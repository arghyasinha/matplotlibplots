```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("NCRB_CII-2020_Table.No-9A.9 (1).csv")
```

```python
data
```

```python
data=data[data.columns[[2,8,17]]]
```

```python
data1=data.loc[0:27,:]
data1
```

```python
X_axis=np.arange(len(data1["State/UT (Col. 2)"]))
plt.bar(X_axis - 0.2, data1["Persons Acquitted - Total - (Col. 17)"], 0.4, label = 'Acquitted')
plt.bar(X_axis + 0.2, data1["Persons Chargesheeted - Total - (Col. 8)"], 0.4, label = 'Chargesheeted')
  
plt.xticks(X_axis,data1["State/UT (Col. 2)"],rotation=90)
plt.xlabel("States")
plt.ylabel("Number of Persons")
plt.title("State-Wise Disposal Of Persons Arrested For Cyber Crime Cases During 2020")
plt.tight_layout()
plt.legend()
plt.savefig("cyb.png",bbox_inches='tight',dpi=300)
plt.show()
```
