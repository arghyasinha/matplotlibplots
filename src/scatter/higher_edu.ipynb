```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("rds2011_12_Table_22.csv")
#data_source= https://data.gov.in/resource/faculty-wise-enrolment-higher-education-1985-86-2011-12
data
```

```python
data=data.loc[[0,1,2,3,4,6,7,8,9],:]
```

```python
data
```

```python
data.drop(list(data.filter(regex = '%')), axis = 1, inplace = True)
```

```python
data
```

```python
TIME=np.array([1985,1995,2001,2005,2006,2007,2008,2009,2010,2011])
TIME.shape
```

```python
np.array(data.loc[0][2:]).shape
```

```python
for i in data.index:
    Y=np.array(data.loc[i][2:])
    plt.scatter(TIME,Y,label=data.loc[i][1],alpha=0.9)
    
plt.legend()
plt.title("Faculty wise enrolment in Higher education in 1985-2011")
plt.tight_layout()
plt.savefig("higher_edu.png",dpi=300)
plt.show()
```
