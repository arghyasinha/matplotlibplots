```python
cur_data1=data[data["CRIME HEAD"]=="TOTAL CRIMES AGAINST STs"].loc[380:407,:]
```

```python
cols=list(cur_data1.columns)
cols.remove('CRIME HEAD')
```

```python
cur_data=cur_data1[cols]
```

```python
tcur=cur_data.set_index("STATE/UT").T
```

```python
tcur.boxplot()
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=8)
plt.tight_layout()
plt.title("Total crimes against STs in the years 2001-2012 among states",)
plt.savefig("bx.png",format='png',bbox_inches='tight',dpi=300)
plt.show()
```
